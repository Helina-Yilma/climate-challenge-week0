```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('../data/ethiopia_clean.csv', parse_dates=['Date'])

# Plot T2M
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['T2M'], label='Avg Temp (2m)')
# Annotation for max
max_temp = df['T2M'].max()
max_date = df.loc[df['T2M'].idxmax(), 'Date']
plt.annotate(f'Max: {max_temp}°C', xy=(max_date, max_temp), arrowprops=dict(facecolor='red'))
plt.title('Monthly Average Temperature (2015-2026)')
plt.show()
```
